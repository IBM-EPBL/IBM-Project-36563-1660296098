```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('/content/forest_fire_1650972473213_1650972490035.webp')
plt.imshow(img)
```

```python
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
img_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(img_gray)
```

```python
img_rgb.shape
```

```python
img_resize = cv2.resize(img_rgb,(500,400))
plt.imshow(img_resize)
```

```python
img_rotate = cv2.rotate(img_resize,cv2.ROTATE_90_CLOCKWISE)
plt.imshow(img_rotate)
```

```python
img_crop = img_resize[175:220,250:300]
plt.imshow(img_crop)
```

```python
img_blur = cv2.GaussianBlur(img_resize,(11,11),cv2.BORDER_ISOLATED)
plt.imshow(img_blur)
```

```python
im=cv2.copyMakeBorder(img_crop,5,5,10,10,cv2.BORDER_CONSTANT)
plt.imshow(im)
```

```python
img_canny = cv2.Canny(img_resize,20,20)
plt.imshow(img_canny)
```

```python
img2 = cv2.imread('/content/forest_fire_1650972473213_1650972490035.webp')
img2_canny = cv2.Canny(img2,150,150)
plt.imshow(img2_canny)
```

```python
cap = cv2.VideoCapture(r'/content/video.mp4')

while True:
    success,frame = cap.read()
    cv2.imshow('frames',frame)
    if cv2.waitKey(0)&0xFF==ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
    
```
